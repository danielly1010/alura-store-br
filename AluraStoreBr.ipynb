### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)


loja.head()
```

```python
# Importação das bibliotecas:
import pandas as pd
import matplotlib.pyplot as plt
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja1["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print("Faturamento por Loja: ")
print("----------------------")
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
for i, loja in enumerate([loja1, loja2, loja3, loja4]):
    faturamento = loja["Preço"].sum()
    print(f"{lojas[i]}: R$ {faturamento:.2f}")
```

```python
#Criar um gráfico barras para visualizar o faturamento:
import matplotlib.pyplot as plt

faturamento_lojas = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ['#493D9E', '#B2A5FF', '#DAD2FF', '#E9A5F1']

for i, valor in enumerate(faturamento_lojas):
    plt.text(i, valor + 10000, f"R$ {valor:,.0f}", ha='center', va='bottom', fontsize=10)

plt.bar(lojas, faturamento_lojas, color=cores)
plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.show()
```

```python
#Gráfico pizza:
import matplotlib.pyplot as plt

faturamento_lojas = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

plt.pie(faturamento_lojas, labels=lojas, autopct='%1.1f%%')
plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.show()
```

# 2. Vendas por Categoria


```python
import pandas as pd

print("Loja 1 - Produtos por Categoria")
print("-------------------------------")
print(loja1["Categoria do Produto"].value_counts())
print("\n")

print("Loja 2 - Produtos por Categoria")
print("-------------------------------")
print(loja2["Categoria do Produto"].value_counts())
print("\n")

print("Loja 3 - Produtos por Categoria")
print("-------------------------------")
print(loja3["Categoria do Produto"].value_counts())
print("\n")

print("Loja 4 - Produtos por Categoria")
print("-------------------------------")
print(loja4["Categoria do Produto"].value_counts())
print("\n")
```

```python

import matplotlib.pyplot as plt
import numpy as np

lojas = [loja1, loja2, loja3, loja4]
categorias = []

for loja in lojas:
    vendas_categoria = loja["Categoria do Produto"].value_counts()
    categorias.extend(vendas_categoria.index.tolist())

categorias = list(set(categorias))

vendas_lojas = []

for loja in lojas:
    vendas_categoria = loja["Categoria do Produto"].value_counts()
    vendas_loja = [vendas_categoria.get(categoria, 0) for categoria in categorias]
    vendas_lojas.append(vendas_loja)

bar_width = 0.2
x = np.arange(len(categorias))

plt.figure(figsize=(10, 6))
for i, vendas_loja in enumerate(vendas_lojas):
    plt.bar(x + i * bar_width, vendas_loja, width=bar_width, label=f"Loja {i+1}")

plt.xlabel('Categoria')
plt.ylabel('Quantidade de Produtos')
plt.title('Vendas por Categoria')
plt.xticks(x + bar_width * (len(lojas) - 1) / 2, categorias, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python


media_avaliacao_loja1 = loja1["Avaliação da compra"].mean()
media_avaliacao_loja2 = loja2["Avaliação da compra"].mean()
media_avaliacao_loja3 = loja3["Avaliação da compra"].mean()
media_avaliacao_loja4 = loja4["Avaliação da compra"].mean()


print("Média de Avaliação das Lojas:")
print("-----------------------------")
print(f"Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Loja 4: {media_avaliacao_loja4:.2f}")

```

# 4. Produtos Mais e Menos Vendidos

```python
lojas = [loja1, loja2, loja3, loja4]

for i, loja in enumerate(lojas):
    print(f"Loja {i+1}:")
    print("------------")
    print("Produtos Mais Vendidos:")
    print(loja["Produto"].value_counts().head(5))
    print("\nProdutos Menos Vendidos:")
    print(loja["Produto"].value_counts().tail(5))
    print("\n")
```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = loja1["Frete"].mean()
frete_medio_loja2 = loja2["Frete"].mean()
frete_medio_loja3 = loja3["Frete"].mean()
frete_medio_loja4 = loja4["Frete"].mean()

print("Frete Médio das Lojas:")
print("-------------------------------")
print(f"Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Loja 4: R$ {frete_medio_loja4:.2f}")

```

```python
import matplotlib.pyplot as plt

lojas = [loja1, loja2, loja3, loja4]
fretes_medios = [loja["Frete"].mean() for loja in lojas]

plt.figure(figsize=(10, 6))
plt.bar(range(1, len(lojas) + 1), fretes_medios)
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')
plt.title('Frete Médio por Loja')
plt.xticks(range(1, len(lojas) + 1), [f"Loja {i}" for i in range(1, len(lojas) + 1)])

for i, valor in enumerate(fretes_medios):
    plt.text(i + 1, valor, f"R${valor:.2f}", ha='center', va='bottom')

plt.show()
```
